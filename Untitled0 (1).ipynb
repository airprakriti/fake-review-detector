```python
from datetime import datetime, timedelta
import random

def generate_review_time(row):
    base_time = datetime.strptime(row['review_date'], '%m/%d/%Y')
    if row['verified_purchase']:
        if random.uniform(0, 1) < 0.62:
            hour = random.randint(6, 23)
        else:
            hour = random.randint(0, 7)
    else:
        if random.uniform(0, 1) < 0.57:
            hour = random.randint(0, 7)
        else:
            hour = random.randint(6, 23)
    return base_time + timedelta(hours=hour, minutes=random.randint(0, 59))

df['review_time'] = df.apply(generate_review_time, axis=1)
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X = ndf1['review_text']
y = ndf1['verified_purchase']
```

```python
X_train, X_test, y_train, y_test = train_test_split(ndf1['review_text'], ndf1['verified_purchase'],test_size=0.4, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vectorizer  = CountVectorizer(stop_words='english')
count_vectorizer.fit(X_train)
print('\nVocabulary: \n', count_vectorizer.vocabulary_)
#max(count_vectorizer.vocabulary_)
```

```python

train_c = count_vectorizer.fit_transform(X_train)
test_c = count_vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
lr1 = LogisticRegression()
lr1.fit(train_c, y_train)
prediction = lr1.predict(test_c)
```

```python
lr_a1 = accuracy_score(y_test, prediction)*100
lr_p1 = precision_score(y_test, prediction)* 100
lr_r1 = recall_score(y_test, prediction)*100
lr_f11 = f1_score(y_test, prediction)*100
```
